```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('udemy_courses (1).csv')
```

```python
data
```

# 1. Check for duplicates and drop them

```python
data = data.drop_duplicates()       # remove duplicate
```

```python
data.duplicated().any()              # check duplicate
```

# 2. Find number of courses per subject

```python
data.head(2)
```

```python
data['subject'].value_counts()
```

### visualize

```python
plt.xlabel('Courses',color = 'g')
plt.ylabel('Number of courses per subject',color = 'g')
plt.title('Find Number of courses per sub', size = 20,color = 'r')
plt.xticks(rotation = 75)

sns.countplot(data['subject'])
plt.show()
```

# 3.  For Which Levels, Udemy Courses Providing The Courses

# 5.Which Course Has More Lectures (Free or Paid)?

```python
data.head(2)
```

```python
data.groupby(['is_paid']).mean()
```
